```python
df = pd.read_csv(r"C:\Users\abhir\Desktop\Masters Analytics\uni_data\geo_data\geo_data.csv")
```

```python
import folium 
map = folium.Map(location=[0, 0], zoom_start=2)
```

```python
for i in range(len(df)):
    University = df["University Name"][i]
    popup = f"{University}"
    point = [df["lat"][i], df["long"][i]]
    folium.Marker(point, popup= popup).add_to(map)
```
